```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# DATASET PARSER!


def parse(year, month, day, hour, minute):
    return year + '-' + month + '-' + day + ' ' + hour + ':' + minute
```

# Data Visualization

```python
# CONCATENATION OF ALL DATAFRAMES!!
twentyall = [twenty10, twenty11, twenty12, twenty13, twenty14,
             twenty15, twenty16, twenty17, twenty18, twenty19]
twentyall = pd.concat(twentyall)
# They are alive
print(twentyall.head())
print(twentyall.tail())
```

## Visualization of a Year

```python
plt.figure(figsize=(16, 9))
twentyall['2019']['GHI'].plot(color='k')
plt.title('GHI during 2019', fontsize=19)
plt.ylabel('GHI in $W m^-2$', fontsize=15)
plt.xlabel('Dates', fontsize=15)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.show()
```

## Visualization of a Month

```python
plt.figure(figsize=(16, 9))
twentyall['2015-5']['GHI'].plot(color='k', x_compat=True)
plt.title('GHI during May 2015', fontsize=19)
plt.ylabel('GHI in $W m^-2$', fontsize=15)
plt.xlabel('Dates', fontsize=15)
plt.xticks(fontsize=13, rotation=60)
plt.yticks(fontsize=13)
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(16,9))
twentyall['GHI'].resample('M').max().plot(color='k')
plt.title('Maximum Irradiance Each Month',fontsize=19)
plt.ylabel('GHI in $W m^-2$', fontsize=15)
plt.xlabel('Dates', fontsize=15)
plt.xticks(fontsize=13, rotation=60)
plt.yticks(fontsize=13)
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(16,9))
plt.plot(twentyall['2019-12']['GHI'].values,'b--',label='Irradiance on December')
plt.plot(twentyall['2019-7']['GHI'].values,'k',label='Irradiance on July')
plt.legend(fontsize=15)
plt.ylim(0,1200)
plt.xlim(0,len(twentyall['2019-12']['GHI'].values))
plt.ylabel('GHI in $W m^-2$', fontsize=15)
plt.xlabel('30 minutes intervals', fontsize=15)
plt.xticks(fontsize=13, rotation=60)
plt.yticks(fontsize=13)
plt.title('Comparison Between Minimum and Maximum Months',fontsize=19)
plt.show()
```

### Checking Stationarity with the Augmented Dickey-Fuller test:

```python
from statsmodels.tsa.stattools import adfuller as adf
X = twentyall['2019']['GHI'].values
result = adf(X)
print( ' ADF Statistic: %.4f ' % result[0])
print( ' p-value: %.4E ' % result[1])
print( ' Critical Values: ' )
for key, value in result[4].items():
    print( ' \t%s: %.4f ' % (key, value))
```

This is an Stationary Time Series.
